## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3,4 ]
y= [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show 
fig.savefig("./images/sample-plot-2.png")
```

```python
# make a plot from a directory
nut_butter_price = {"almon butter": 10,
                   "peanut butter": 8,
                   "cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
ax.set(title="Hamed butter store",
      xlabel="Butter type",
      ylabel="Price");
```

```python
# Horizontal bar plor 
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()))
```

```python
# Make some data for histogram and plot it
x= np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.randn(100), np.random.randn(100))
ax3.bar(nut_butter_price.keys(), nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randn(100), np.random.randn(100))
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values())
ax[1, 1].hist(np.random.randn(1000));
```

```python
NUMS = 1000
ts = pd.Series(np.random.randn(NUMS),
              index=pd.date_range(start='01/01/2010', periods=NUMS))
ts_row = ts
ts_sum = ts.cumsum()

fig, (ax1, ax2) = plt.subplots(nrows=1,
                              ncols=2,
                              figsize=(10, 5))
ax1.plot(ts_row)
ax2.plot(ts_sum);
```

```python
# Lets try another dataset
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist();
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 20), subplots=True);
```

```python
heart_disease.head()
```

```python
type(heart_disease["age"][0])
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(x="age",
            y="chol",
            kind="scatter",
            c="target",
            figsize = (10, 5),
            colormap='viridis',
            xlim=[45, 80]);
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize= (10, 5))
over_50.plot(x="age",
            y="chol",
            ax=ax,
            kind="scatter",
            c="target",
            colormap="viridis",
            xlim=[45, 80]);
```

```python
# OO method from the scratch
fig, ax = plt.subplots(figsize = (10, 5))

# Plot the data
scatter = ax.scatter(x= over_50["age"],
                     y= over_50["chol"],
                     c= over_50["target"])

# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="cholesterol");

# Add legend
ax.legend(*scatter.legend_elements(), title="target")

# Add horizontal line to plot
ax.axhline(over_50["chol"].mean(),
           c='blue',
           ls='--',
           lw=2.0)
```
