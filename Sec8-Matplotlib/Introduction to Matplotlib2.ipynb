## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3,4 ]
y= [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show 
fig.savefig("./images/sample-plot-2.png")
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.randn(100), np.random.randn(100))
ax3.bar(nut_butter_price.keys(), nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randn(100), np.random.randn(100))
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values())
ax[1, 1].hist(np.random.randn(1000));
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize= (10, 5))
over_50.plot(x="age",
            y="chol",
            ax=ax,
            kind="scatter",
            c="target",
            colormap="viridis",
            xlim=[45, 80]);
```

```python
# OO method from the scratch
fig, ax = plt.subplots(figsize = (10, 5))

# Plot the data
scatter = ax.scatter(x= over_50["age"],
                     y= over_50["chol"],
                     c= over_50["target"])

# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="cholesterol");

# Add legend
ax.legend(*scatter.legend_elements(), title="target")

# Add horizontal line to plot
ax.axhline(over_50["chol"].mean(),
           c='blue',
           ls='--',
           lw=2.0)
```

## Adding another plot to existing style one

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              sharex=True,
                              figsize = (10, 8))

# Add data for ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Taget")

# Set a mean line
ax0.axhline(y=over_50["chol"].mean(),
            color = "b",
            ls="--",
            lw=2.0,
           label="Average")

# Add data for ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")


# Set a mean line
ax1.axhline(y=over_50["thalach"].mean(),
            c="b",
            ls="--",
            lw=2,
            label="Average")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight="bold");
```

## Changing the xlim and ylim

```python
## Before the change (we've had color updates)

fig, (ax0, ax1) = plt.subplots(figsize= (10, 10),
                              nrows=2,
                              ncols=1,
                              sharex=True)

scatter = ax0.scatter(over_50["age"],
                     over_50["chol"],
                     c= over_50["target"],
                     cmap="winter")

ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Set a mean H-line
ax0.axhline(over_50["chol"].mean(),
            ls="--",
            lw=2,
            c="r")

ax0.legend(*scatter.legend_elements(), title="target")

# Axis 1, 1
scatter = ax1.scatter(over_50["age"],
                     over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Set mean line
ax1.axhline(over_50["thalach"].mean(),
            ls="--",
            lw=2,
            c="r",
            label="Average")

ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
## After adding in different x & y limitations
fig, (ax0, ax1) = plt.subplots(figsize= (10, 10),
                              nrows=2,
                              ncols=1,
                              sharex=True)

scatter = ax0.scatter(over_50["age"],
                     over_50["chol"],
                     c= over_50["target"],
                     cmap="winter")

ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Set the x-axis
ax0.set_xlim([50, 80])

# Set a mean H-line
ax0.axhline(over_50["chol"].mean(),
            ls="--",
            lw=2,
            c="r")

ax0.legend(*scatter.legend_elements(), title="target")

# Axis 1, 1
scatter = ax1.scatter(over_50["age"],
                     over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Set the y axis
ax1.set_ylabel([60, 200])

# Set mean line
ax1.axhline(over_50["thalach"].mean(),
            ls="--",
            lw=2,
            c="r",
            label="Average")

ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Create the figure
fig, ax = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10, 10),
                      sharex=True)

# Plot data about cholesterol
scatter1 = ax[0].scatter(x= over_50["age"],
                         y= over_50["chol"],
                         c= over_50["target"],
                        cmap= "winter")

# plot the second data thalach
scatter2 = ax[1].scatter(x= over_50["age"],
                         y= over_50["thalach"],
                         c= over_50["target"],
                        cmap="winter")

# Customize the plot 
ax[0].set(title = 'Heart Disease and cholesterol level',
          ylabel = 'Cholesterol')
ax[1].set(title = 'Heart disease and max heart rate', 
          xlabel = 'Age', 
          ylabel = 'Max heart rate')

# Change the x axis
ax[0].set_xlim([50, 80])
ax[1].set_xlim([50, 80])

# Change the y axis
ax[0].set_ylim([100, 600])
ax[1].set_ylim([60, 200])

# Set legends
ax[0].legend(*scatter1.legend_elements(), title = 'Target')
ax[1].legend(*scatter2.legend_elements(), title = 'Target')

# Add mean line 
ax[0].axhline(y = over_50['chol'].mean(),
              ls = '--',
              lw= 1.5, 
              color = 'k')
ax[1].axhline(y = over_50['thalach'].mean(),
              ls = '--',
             color = 'g',
             lw = 1.5)

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize = 14, fontweight="bold");
```
