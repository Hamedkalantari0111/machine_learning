## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Sample Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## making figures with NumPy arrays
We want:
* Line plot
* scatter plot
* Bar plot
* Histogram
* subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# use the data to create the scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot rom a dictionary
nut_butter_price = {"Almon butter": 10,
                    "Peanut buuter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
ax.set(title = "Hamed's buuter store",
       xlabel = "Butter type",
       ylabel = "Price");
```

```python
# Horizontal bar plor 
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(nut_butter_price.keys(), nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(100), np.random.random(100))
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values())
ax[1, 1].hist(np.random.randn(1000));
```
