## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Sample Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot rom a dictionary
nut_butter_price = {"Almon butter": 10,
                    "Peanut buuter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
ax.set(title = "Hamed's buuter store",
       xlabel = "Butter type",
       ylabel = "Price");
```

```python
# Horizontal bar plor 
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(nut_butter_price.keys(), nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(100), np.random.random(100))
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly, use the pyplot method. 
* When plotting something more anvanced, use the OO method.

```python
heart_disease.head()
```

```python
type(heart_disease['age'][0])
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
over_50.head()
```

```python
# Pyplot mehtod
over_50.plot(kind='scatter', 
             x='age',
             y='chol',
             c='target');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10, 5))
over_50.plot(kind='scatter', 
             x='age',
             y='chol',
             c='target',
             ax=ax,
             xlim=[45, 100]);
```

```python
# OO method from the scratch
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target']);

# customize the plot 
ax.set(title = 'Heart disease and cholesterol levels',
       xlabel = 'Age',
       ylabel = 'cholesterol');

# Add legend
ax.legend(*scatter.legend_elements(), title='target');

# Add horizontal line to plot
ax.axhline(over_50['chol'].mean(),
           c='blue',
           ls='--',
           lw=2.0);
```

```python
heart_disease.head()
```

```python
# create the figure
fig, ax = plt.subplots(nrows=2,
                       ncols=1,
                       figsize=(10, 10), 
                       sharex = True)

# Plot data about cholesterol
scatter1 = ax[0].scatter(x=over_50['age'],
                         y=over_50['chol'],
                         c=over_50['target']);

# plot the second data thalach
scatter2 = ax[1].scatter(x=over_50['age'],
                         y=over_50['thalach'],
                        c=over_50['target']);

# Customize the plot 
ax[0].set(title = 'Heart Disease and cholesterol level',
          ylabel = 'Cholesterol')
ax[1].set(title = 'Heart disease and max heart rate', 
          xlabel = 'Age', 
          ylabel = 'Max heart rate')

# Set legends
ax[0].legend(*scatter1.legend_elements(), title = 'Target')
ax[1].legend(*scatter2.legend_elements(), title = 'Target');

# Add mean line 
ax[0].axhline(y = over_50['chol'].mean(),
              ls = '--',
              lw= 1.5, 
              color = 'k')
ax[1].axhline(y = over_50['thalach'].mean(),
              ls = '--',
             color = 'g',
             lw = 1.5);

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize = 14);
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from the scratch
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'], 
                     cmap = 'winter');

# customize the plot 
ax.set(title = 'Heart disease and cholesterol levels',
       xlabel = 'Age',
       ylabel = 'cholesterol');

# Add legend
ax.legend(*scatter.legend_elements(), title='target');

# Add horizontal line to plot
ax.axhline(over_50['chol'].mean(),
           c='blue',
           ls='--',
           lw=2.0);
```

```python
# Customizing the x, y axes limitatons

# create the figure
fig, ax = plt.subplots(nrows=2,
                       ncols=1,
                       figsize=(10, 10), 
                       sharex = True)

# Plot data about cholesterol
scatter1 = ax[0].scatter(x=over_50['age'],
                         y=over_50['chol'],
                         c=over_50['target'], 
                         cmap = 'winter');

# plot the second data thalach
scatter2 = ax[1].scatter(x=over_50['age'],
                         y=over_50['thalach'],
                        c=over_50['target'], 
                         cmap = 'winter');

# Customize the plot 
ax[0].set(title = 'Heart Disease and cholesterol level',
          ylabel = 'Cholesterol')
ax[1].set(title = 'Heart disease and max heart rate', 
          xlabel = 'Age', 
          ylabel = 'Max heart rate')

# Change the x axis
ax[0].set_xlim([50, 80])
ax[1].set_xlim([50, 80])

# Change the y axis
ax[0].set_ylim([100, 600])
#ax[1].set_ylim([60, 200])

# Set legends
ax[0].legend(*scatter1.legend_elements(), title = 'Target')
ax[1].legend(*scatter2.legend_elements(), title = 'Target');

# Add mean line 
ax[0].axhline(y = over_50['chol'].mean(),
              ls = '--',
              lw= 1.5, 
              color = 'k')
ax[1].axhline(y = over_50['thalach'].mean(),
              ls = '--',
             color = 'g',
             lw = 1.5);

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize = 14);
```
